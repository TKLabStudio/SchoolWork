```
#tensorflow 2.0
import tensorflow as tf
mnist = tf.keras.datasets.mnist  #mnist model寫法

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
model = tf.keras.models.Sequential([   #創建model
  tf.keras.layers.Flatten(input_shape=(28, 28)), #28*28二維矩陣  轉為  一維矩陣
  tf.keras.layers.Dense(1024, activation='relu'),
  tf.keras.layers.Dropout(0.2),  
  tf.keras.layers.Dense(10, activation='softmax') #0~9 就是十個  softmax比較好
])
model.summary() #把你的layer顯示出來
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=15)

model.evaluate(x_test,  y_test, verbose=2) #結果
```

```
import tensorflow as tf


(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data() #train跟test的寫法

model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(32, 32,3)),
  tf.keras.layers.Dense(1024, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(1024, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='softmax')
])
model.summary() #layer顯示
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=15)

model.evaluate(x_test,  y_test, verbose=2) #結果
```

```
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
import numpy as np
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
model = Sequential()
model.add(Dense(1024, activation='relu', input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))
model.summary()
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
history = model.fit(x_train, y_train,
                    epochs=15,
                    verbose=1,
                    validation_data=(x_test, y_test))
model.evaluate(x_test,  y_test, verbose=2)
```

```
#keras mnist
import keras #import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
import numpy as np
from keras.layers import Conv2D  # Convolution Operation
from keras.layers import MaxPooling2D # Pooling
from keras.layers import Flatten
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1).astype('float32')#進行資料預處理
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1).astype('float32')  
x_train /= 255
x_test /= 255
y_train = keras.utils.to_categorical(y_train, 10)#One-Hot Encoding
y_test = keras.utils.to_categorical(y_test, 10)
model = Sequential()#model
model.add(Conv2D(128, (3, 3), input_shape = (28, 28,1), activation = 'relu'))#convolution
model.add(MaxPooling2D(pool_size = (2, 2)))#pooling
model.add(Dropout(0.35))
model.add(Conv2D(128,  (3, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.35))
model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))
model.summary()
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
history = model.fit(x_train, y_train,
                    epochs=15,
                    verbose=1)
model.evaluate(x_test,  y_test, verbose=2)
```
